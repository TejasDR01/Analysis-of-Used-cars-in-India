# Reading the file and Filling the NAN values

# Removing Outliers

```python
def remove_outliers(array):
    Q1=array.quantile(0.25)
    Q3=array.quantile(0.75)
    IQR=Q3-Q1
    t=(array >=Q1-1.5*IQR)&(array <=Q3+1.5*IQR)
    return t
kd=remove_outliers(temp["Kilometers_Driven"])
Eng=remove_outliers(temp["Engine(CC)"])
Mil=remove_outliers(temp["Mileage(kmpl)"])
Pow=remove_outliers(temp["Power(bph)"])
Price=remove_outliers(temp["Price(L)"])
temp=temp.loc[kd]
temp=temp.loc[Eng]
temp=temp.loc[Mil]
temp=temp.loc[Pow]
temp=temp.loc[Price]
temp
```

# Normalization

```python
def normalize(array,mean,std):
    f=[]
    for i in array:
        f.append((i-mean)/std)
    return f
kd_norm=normalize(temp["Kilometers_Driven"],temp["Kilometers_Driven"].mean(),temp["Kilometers_Driven"].std())
Eng_norm=normalize(temp["Engine(CC)"],temp["Engine(CC)"].mean(),temp["Engine(CC)"].std())
Mil_norm=normalize(temp["Mileage(kmpl)"],temp["Mileage(kmpl)"].mean(),temp["Mileage(kmpl)"].std())
Pow_norm=normalize(temp["Power(bph)"],temp["Power(bph)"].mean(),temp["Power(bph)"].std())
Price_norm=normalize(temp["Price(L)"],temp["Price(L)"].mean(),temp["Price(L)"].std())
df=pd.DataFrame()
df["Kilometers_Driven"]=kd_norm
df["Engine(CC)"]=Eng_norm
df["Mileage(kmpl)"]=Mil_norm
df["Power(bph)"]=Pow_norm
df["Price(L)"]=Price_norm
df
```

# Hypothesis Testing

```python
from scipy.stats import norm
from math import sqrt
def two_side_hypo(smean,popmean,std,samplesize,alpha):
    actual_z=abs(norm.ppf(alpha/2))
    hypo_z=abs((smean-popmean)/(std/sqrt(samplesize)))
    print("actual_z: "+str(actual_z))
    print("hypo_z: "+str(hypo_z))
    if(hypo_z>=actual_z):
        return True
    else:
        return False
    
def one_side_hypo(smean,popmean,std,samplesize,alpha):
    actual_z=abs(norm.ppf(alpha))
    hypo_z=abs((smean-popmean)/(std/sqrt(samplesize)))
    print("actual_z: "+str(actual_z))
    print("hypo_z: "+str(hypo_z))
    if(hypo_z>=actual_z):
        return True
    else:
        return False
print("For column 'Kilometers_Driven'")
print("")
print("we draw random sample of size 60")
kilodriv=[i for i in temp["Kilometers_Driven"]]
sample=[kilodriv[i] for i in range(100,400) if i%5==0]
smean=round(np.mean(sample),2)
std=round(np.std(temp["Kilometers_Driven"]),2)
print("sample mean: "+str(smean))
print("since it is drawn from a normal population we use z-test")
print("")
alpha=0.05
popmean=57000
samplesize=60
print("Ho : μ = 57000")
print("Ha : μ != 57000")
print("")
if two_side_hypo(smean,popmean,std,samplesize,alpha):
    print("We reject the null hypothesis")
else:
    print("we fail to reject the null hypothesis")
```

```python
print("For column 'Mileage(kmpl)'")
print("")
print("we draw random sample of size 25")
mile=[i for i in temp["Mileage(kmpl)"]]
sample=[mile[i] for i in range(300,425) if i%5==0]
smean=round(np.mean(sample),2)
std=round(np.std(temp["Mileage(kmpl)"]),2)
print("sample mean: "+str(smean))
print("since it is drawn from a normal population\nand the sample size is small but\nwe have the population std we go with the z-test")
print("")
alpha=0.05
popmean=20
samplesize=25
print("Ho : μ >= 20")
print("Ha : μ < 20")
print("")
if one_side_hypo(smean,popmean,std,samplesize,alpha):
    print("We reject the null hypothesis")
else:
    print("we fail to reject the null hypothesis")
```

# Correlation

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error as RMSE
def correlation_coefficient(d1,d2):
    data1=[i for i in d1]
    data2=[i for i in d2]
    x=np.mean(data1)
    y=np.mean(data2)
    xy=0
    yy=0
    xx=0
    if(len([data1])!=len([data2])):
        return("lengths are not same")
    for i in range(0,len(data1)):
        xy+=(data1[i]-x)*(data2[i]-y)
        xx+=(data1[i]-x)**2
        yy+=(data2[i]-y)**2
    co_coef=xy/(sqrt(xx*yy))
    return co_coef
def linear_regression(d1,d2):
    x=[i for i in d1]
    y=[i for i in d2]
    x=[[i] for i in x]
    regr = linear_model.LinearRegression()
    regr.fit(x,y)
    y_preds = regr.predict(x)
    print("coefficients: \n"+str(regr.coef_))
    print("RMSE: %.2f"% RMSE(y,y_preds))
    plt.scatter(x,y, color='m')
    plt.plot(x,y_preds, color='g')
    plt.show()
print("the correlation coefficient is : "+str(correlation_coefficient(temp["Engine(CC)"],temp["Power(bph)"])))
plt.xlabel("Engine(CC)")
plt.ylabel("Power(bph)")
plt.scatter(temp["Engine(CC)"],temp["Power(bph)"])
plt.show()
linear_regression(temp["Engine(CC)"],temp["Power(bph)"])
print("as you can see the correlation coffecient is about 0.7\nand the least squares line also shows a\npositive relationship between the variables")
```
